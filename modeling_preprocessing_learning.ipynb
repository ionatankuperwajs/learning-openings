# Preprocessing the data for model fitting

```python
import pandas as pd
import pickle
import numpy as np
```

## Queen's Gambit

```python
# Load the csv data of QG games (October 2020 - March 2021)
direc = '../../Data/Learning/queens_gambit/'

df = pd.concat([pd.read_csv(direc + 'df_{:d}_{:02d}.csv'.format(2020 + j//12,j%12+1),
                            usecols=['black','fourthmove','result'])
                for j in range(9,15)],ignore_index=True)
```

```python
# Load the pkl data of unique players (March 2020 - September 2020)
players = []
for i in range(3,9):
    with open(direc + 'QG_players_2020_{:02d}.pkl'.format(i),'rb') as f:
        players = pickle.load(f)
s = set(np.array(players).flatten())

# Filter out players who have played the QG in this period
df = df[df['black'].apply(lambda p: not p in s)].reset_index(drop=True)
```

```python
def get_ordered_unique_values(a):
    x,c = np.unique(a,return_counts=True)
    x = x[np.argsort(c)[::-1]]
    c = np.sort(c)[::-1]
    return x,c
```

```python
# Add accept (1), decline (either the Slav Defense or QG Declined, 2), other to df
df['fourthcode'] = df['fourthmove'].map({'d5c4' : 1, 'e7e6' : 2, 'c7c6' : 2}).fillna('O')
```

```python
# Create a binary dataframe of just accept/decline
df_binary = df[(df['fourthcode'].isin([1,2])) & (~df['result'].isna())].reset_index(drop=True)
names,ind = np.unique(df_binary['black'],return_inverse=True)
name_to_ind = dict(zip(names,ind))

df_binary = df_binary.drop('fourthmove', axis=1)
```

```python
# Map losses to -1 instead of 0 and draws to 0 instead of 0.5
df_binary['result'] = df_binary['result'].map({0.0 : -1, 1.0: 1, 0.5 : 0})
```

```python
# Change to integers and swap columns
df_binary['result'] = df_binary['result'].astype('Int64')
df_binary['fourthcode'] = df_binary['fourthcode'].astype('Int64')
df_binary = df_binary[['black', 'fourthcode', 'result']]
```

```python
# Label unique users starting at 1
df_binary['black'] = pd.factorize(df_binary['black'])[0] + 1
df_binary
```

```python
# Save to csv
df_binary.to_csv('modeling_QG.csv', index=False)
```
